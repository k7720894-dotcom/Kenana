```python

PC1 = [
    57, 49, 41, 33, 25, 17, 9,
    1,  58, 50, 42, 34, 26, 18,
    10, 2,  59, 51, 43, 35, 27,
    19, 11, 3,  60, 52, 44, 36,
    63, 55, 47, 39, 31, 23, 15,
    7,  62, 54, 46, 38, 30, 22,
    14, 6,  61, 53, 45, 37, 29,
    21, 13, 5,  28, 20, 12, 4
]

PC2 = [
    14, 17, 11, 24, 1,  5,
    3,  28, 15, 6,  21, 10,
    23, 19, 12, 4,  26, 8,
    16, 7,  27, 20, 13, 2,
    41, 52, 31, 37, 47, 55,
    30, 40, 51, 45, 33, 48,
    44, 49, 39, 56, 34, 53,
    46, 42, 50, 36, 29, 32
]


SHIFT_SCHEDULE = [1, 1, 2, 2, 2, 2, 2, 2, 1, 2, 2, 2, 2, 2, 2, 1]


def key_to_bits(key_bytes):
    bits = []
    for b in key_bytes:
        for i in range(8):
            bits.append((b >> (7-i)) & 1)
    return bits


def left_shift(bits, n):
    return bits[n:] + bits[:n]


def generate_subkeys(key_bytes):
    bits = key_to_bits(key_bytes) 
    
    permuted = [bits[i-1] for i in PC1]


    C = permuted[:28]
    D = permuted[28:]

    subkeys = []

    for shift in SHIFT_SCHEDULE:
        
        C = left_shift(C, shift)
        D = left_shift(D, shift)
    
        CD = C + D
        
        subkey = [CD[i-1] for i in PC2]
        subkeys.append(subkey)

    return subkeys
```
