```python
def split_into_pairs(plaintext):
    
    processed = ""
    i = 0

    while i < len(plaintext):
        char1 = plaintext[i]
        char2 = ''

        if i + 1 < len(plaintext):
            char2 = plaintext[i + 1]
            if char1 == char2:
                processed += char1 + 'x'
                i += 1
            else:
                processed += char1 + char2
                i += 2
        else:
            
            processed += char1
            i += 1

    
    if len(processed) % 2 != 0:
        processed += 'z'

    return processed
```

```python
def remove_duplicate_letters(key):
    
    seen = set()
    cleaned_key = ""

    for char in key:
        if char not in seen:
            cleaned_key += char
            seen.add(char)

    return cleaned_key
```

```python
def find_position(matrix, char):
    
    for i, row in enumerate(matrix):
        for j, c in enumerate(row):
            if c == char:
                return [i, j] 
    return None 
```
