```python
def clean_text(text):
    return text.replace(" ","")
```

```python
def built_key(key,plaintext):
    plaintext=plaintext.lower()
    key=key.lower()
    while len(key)<len(plaintext):
        key+=key
    return key[:len(plaintext)]
```

```python
def encrypt_vigenere(plaintext,key):
    plaintext=clean_text(plaintext)
    plaintext=plaintext.lower()
    key_stream=built_key(key,plaintext)
    enrypted=""
    for p_char,k_char in zip(plaintext,key_stream):
        p=ord(p_char)-ord('a')
        k=ord(k_char)-ord('a')
        c=(p+k) %26
        encrypted+=chr(ord('A') + c)
    return encrypted
```

```python
def decrypt_vigenere(ciphertext,key):
    ciphertext=clean_text(ciphertext)
    key_stream=(key,ciphertext,lower())
    decrypted="" 
    for c_char,k_char in zip(ciphertext,key_stream):
        p=ord(c_char)-ord('A')
        k=ord(k_char)-ord('a')
        c=(c-k) %26
        encrypted+=chr(ord('a') + p)
    return encrypted
```
