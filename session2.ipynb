```python
from math import gcd
#حساب المقلوب الضربي
def mod_inverse(k):
    if gcd(k,26) != 1:
        return None
    for i in range (26):
        if(k*i) % 26 ==1:
            return i
    return None


def decrypt_mult(text,key):
    decrypted=""
    inv=mod_inverse(key)
    if inv is None:
        return "invalid key"
    for char in text:
        c=ord(char)-ord('A')
        p=(c*inv)%26
        decrypted += chr(ord('a')+p)
    return decrypted
```

```python
def bruteforce_multi(ciphertext):
    for key in range(1,26):
        if gcd(key,26) !=1:
            continue
        decrypted=decrypt_mult(ciphertext,key)
        print(f"key {key:2d}:{decrypted}")
```
