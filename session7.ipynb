```python
def rc4_ksa(key):
    
    key_length = len(key)
    S = list(range(256))
    j = 0

    for i in range(256):
        j = (j + S[i] + key[i % key_length]) % 256
        S[i], S[j] = S[j], S[i]  

    return S
```

```python
def rc4_prga(S, n_bytes):
    i = 0
    j = 0
    key_stream = []

    for _ in range(n_bytes):
        i = (i + 1) % 256
        j = (j + S[i]) % 256
        S[i], S[j] = S[j], S[i]
        K = S[(S[i] + S[j]) % 256]
        key_stream.append(K)

    return key_stream
```

```python
def rc4_encrypt(plaintext_bytes, key_bytes):
    
    S = rc4_ksa(key_bytes)
    key_stream = rc4_prga(S, len(plaintext_bytes))
    ciphertext = [p ^ k for p, k in zip(plaintext_bytes, key_stream)]
    return ciphertext
```
