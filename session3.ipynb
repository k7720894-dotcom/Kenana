```python
def built_key(key,plaintext):
    plaintext=plaintext.lower()
    key=key.lower()
    while len(key)<len(plaintext):
        key+=plaintext[len(key)]
        return key
```

```python
def encrypt_autokey(plaintext,key):
    plaintext=plaintext.lower()
    key_stream=built_key(key,plaintext)
    encrypted="" 
    for p_char,k_char in zip(plaintext,key_stream):
        p=ord(p_char)-ord('a')
        k=ord(k_char)-ord('a')
        c=(p+k) %26
        encrypted+=chr(ord('A') + c)
    return encrypted

    
```

```python
def decrypt_autokey(ciphertext,key):
    key=key.lower()
    decrypted=""
    key_stream=key
    for i,c_char in enumerate(ciphertext):
        c=ord(c_char)-ord('A')
        k=ord(key_stream[i] - ord('a'))
        p=(c-k)%26
        p_char=chr(ord('a')+p)
        decrypted+=p_char
    return decrypted
```
