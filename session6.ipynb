```python
def create_adfgvx_matrix(key):
    
    key = key.lower()
    seen = set()
    matrix_chars = []

    for char in key:
        if char.isalnum() and char not in seen:
            matrix_chars.append(char)
            seen.add(char)

    for char in "abcdefghijklmnopqrstuvwxyz0123456789":
        if char not in seen:
            matrix_chars.append(char)
            seen.add(char)

    matrix = [matrix_chars[i*6:(i+1)*6] for i in range(6)]
    return matrix
```

```python
def find_position(matrix, char):
    
    for i, row in enumerate(matrix):
        for j, c in enumerate(row):
            if c == char:
                return [i, j]
    return None
```

```python
def encrypt_adfgvx(plaintext, key):
    
    matrix = create_adfgvx_matrix(key)
    labels = ['A', 'D', 'F', 'G', 'V', 'X']
    ciphertext = ""

    plaintext = plaintext.lower()

    for char in plaintext:
        if char.isalnum():  
            row, col = find_position(matrix, char)
            ciphertext += labels[row] + labels[col]
        else:
            continue  
    return ciphertext 
```
